```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

app_clean_df=application_df.drop(columns= {"EIN", "NAME"})
app_clean_df.head(2)
```

 My target variable is is_successful
 the remaining variables are my features.


```python
# Determine the number of unique values in each column.
app_clean_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = app_clean_df['APPLICATION_TYPE'].value_counts()
app_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_counts[app_counts < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    app_clean_df['APPLICATION_TYPE'] = app_clean_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_clean_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = app_clean_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    app_clean_df['CLASSIFICATION'] = app_clean_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
app_clean_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_clean_df = pd.get_dummies(app_clean_df)
app_clean_df.head()
```

```python
app_clean_df.columns
```

```python
# Split our preprocessed data into our features and target arrays
x = app_clean_df.drop(columns=['IS_SUCCESSFUL']).values
y = app_clean_df['IS_SUCCESSFUL'].values
```

```python
x[0]
```

```python
y
```

```python
# split into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
len(x_train[0])
```

```python
x_test[0]
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
print(x_train_scaled[0])
```

```python
# Create a keras sequentila model
nn_model = tf.keras.models.Sequential()
```

```python
# Add first dense layer including the input layer (hidden layer)
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim =43))
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu"))
```

```python
# output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn_model.fit(x_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('alphabet_soup.h5')
```
